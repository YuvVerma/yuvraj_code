```python
# Calculate the average open and close value per Ticker
average_value = data.groupby("Ticker")[['Open','Close']].mean()
```

```python
# Plotting
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
```

```python
# Bar Plot
average_value.plot(kind="bar",stacked=False, figsize=(10,6), colormap="viridis")
plt.title("Average Open and Close Value per Ticker")
plt.xlabel("Ticker")
plt.ylabel("Average Value")
plt.xticks(rotation=0)
plt.legend(title="Value Type")

# Plot Show
plt.show()
```

```python
# select only numerical features for correlation analysis
numerical_data = data.select_dtypes(include=np.number)
correlation_matrix = numerical_data.corr()

# Display the correlation matrix.
print(correlation_matrix)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('correlation Matrix Heatmap ')
plt.show()
```

**Stock Price Correlation:**

 A correlation matrix heatmap revealed the relationships between adjusted closing prices of different stocks. The analysis used a pivot table with 'Date' as the index to perform the correlation analysis across the time dimension.

```python
# The distribution of closing price to understand their range and frequency.
plt.hist(data['Close'], bins=20)
plt.xlabel('closing price')
plt.ylabel('frequency')
plt.title('closing price distribution')
plt.show()
```

```python
# The cumulative volume traded over time to obverse any trands or spikes.
ticker_volume = data.groupby('Ticker')['Volume'].sum()    # Added parentheses to call the sum function
ticker_volume.plot(kind='bar')
plt.xlabel('Ticker')
plt.ylabel('Total volume')
plt.title('totol Volume by Ticker')
```

```python
# Exploring the relationship between volume and closing prices to identify any relation
plt.scatter(data['Volume'],data['Close'])
plt.xlabel('Volume')
plt.ylabel('Closing price')
plt.title('volume vs closing price')
plt.show()
```

**Volume and Price Relationship**:

Scatter plots illustrated the connection between daily trading volume and daily returns for each stock, helping to visualize the impact of volume on price movements.
